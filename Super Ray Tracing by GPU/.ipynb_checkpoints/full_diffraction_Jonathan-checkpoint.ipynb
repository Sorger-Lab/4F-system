```python
from numba import jit
import scipy
import numpy as np
import time
import matplotlib
import matplotlib.pyplot as plt
```

```python


@jit(nopython=True)
def numba_meshgrid(a, b, indexing_xy=True):
    outA = np.zeros((len(a),len(b)))
    outB = np.zeros((len(a),len(b)))
    if indexing_xy:
        for x in range(0,len(a)):
            for y in range(0,len(b)):
                outA[y,x] = np.float(a[x])
                outB[y,x] = np.float(b[y])
    else:
        for x in range(0,len(a)):
            for y in range(0,len(b)):
                outA[x,y] = np.float(a[x])
                outB[x,y] = np.float(b[y])
    return outA,outB

@jit(nopython=True)
def distMat(px, py, X, Y, z):
    return(np.sqrt((X - px)**2 + (Y - py)**2 + z**2))

def propFull(u1, L, lambda_0, z):
    # full propgation of the Rayleighâ€“Sommerfeld solution (Eq 4.12)
    (M,N)=np.shape(u1)
    L = float(L)
    lambda_0 = float(lambda_0)
    z = float(z)
    M = float(M)
    N = float(N)
    dx = float(L)/float(M)
    k = 2*np.pi/lambda_0
    fx = np.arange(-1/(2*dx),1/(2*dx),1/L)
    (FX,FY) = np.meshgrid(fx,fx)
    uo = np.zeros(np.shape(u1))
    for oi_x in range(0,int(M)):
        for oi_y in range(0,int(N)):
            sums = 0
            r01 = distMat(FX[oi_x,oi_y], FY[oi_x,oi_y], FX, FY, z)
            uo[oi_x, oi_y] = (z/(1j*lambda_0))*np.sum(u1*np.exp(1j*k*r01/r01**2))
    return uo

def propIR(u1, L, lambda_0, z):
    # propagation transfer function assumes same x and y side lengths and uniform sampling
    # u1 - source plane field
    # L - source and observation plane side lengths
    # lambda_0 wavelength
    # z - propagation distance
    # u2 - observation plane field    
    (M,N)=np.shape(u1)
    L = float(L)
    lambda_0 = float(lambda_0)
    z = float(z)
    M = float(M)
    N = float(N)
    dx = float(L)/float(M)
    k = 2*np.pi/lambda_0
    x = np.linspace(-L/2,L/2,int(L/dx))
    print(len(x))
    (X,Y) = np.meshgrid(x,x)
    h = 1/(1j*lambda_0*z)*np.exp(1j*k/(2*z)*(X**2+Y**2))
    H = np.fft.fft2(np.fft.fftshift(h))*dx**2
    U1 = np.fft.fft2(np.fft.fftshift(u1))
    U2 = H*U1
    u2 = np.fft.ifftshift(np.fft.fft2(U2))
    return u2
    
def propTF(u1, L, lambda_0, z):
    # propagation transfer function assumes same x and y side lengths and uniform sampling
    # u1 - source plane field
    # L - source and observation plane side lengths
    # lambda_0 wavelength
    # z - propagation distance
    # u2 - observation plane field
    (M,N)=np.shape(u1)
    L = float(L)
    lambda_0 = float(lambda_0)
    z = float(z)
    M = float(M)
    N = float(N)
    dx = float(L)/float(M)
    k = 2*np.pi/lambda_0
    fx = np.arange(-1/(2*dx),1/(2*dx),1/L)
    (FX,FY) = np.meshgrid(fx,fx)
    H = np.exp(-1j*np.pi*lambda_0*z*(FX**2.0+FY**2.0))
    H = np.fft.fftshift(H)
    U1 = np.fft.fft2(np.fft.fftshift(u1))
    U2 = H*U1
    u2 = np.fft.ifftshift(np.fft.fft2(U2))
    return u2

def propFF(u1,L1,lambda_0,z):
    # propagation - Fraunhofer pattern assumes unifom sampling
    # u1 - source plane field
    # L1 - source plane side length
    # lambda_0 wavelngth
    # z - propagation distance
    # L2 - oversvation plane side length
    # u2 oversvation plane field
    (M,N)=np.shape(u1)
    L1 = float(L1)
    lambda_0 = float(lambda_0)
    z = float(z)
    M = float(M)
    N = float(N)
    dx1 = float(L1)/float(M)
    k = 2*np.pi/lambda_0
    L2 = lambda_0*z/dx1
    dx2 = lambda_0*z/L1
    x2 = np.linspace(-L2/2,L2/2,int(L2/dx2))
    (X2,Y2) = np.meshgrid(x2,x2)
    c = 1/(1j*lambda_0*z)*np.exp(1j*k/(2*z)*(X2**2+Y2**2))
    u2=c*np.fft.fftshift(np.fft.fft2(np.fft.fftshift(u1)))*(dx1**2)
    return u2

def tilt(uin,L,lambda_0,alpha,theta):
    # tilt phasefront
    # uniform sampling assumed
    # uin - input field
    # L - side length
    # lambda_0 - wavelength
    # alpha - tilt angle
    # theta - rotation angle
    (M,N)=np.shape(uin)
    L = float(L)
    lambda_0 = float(lambda_0)
    alpha = float(alpha)
    theta = float(theta)
    M = float(M)
    N = float(N)
    dx = float(L)/float(M)
    k = 2.0*np.pi/lambda_0
    x = np.linspace(-L/2,L/2,int(L/dx))
    (X,Y) = np.meshgrid(x,x)
    uout = uin*np.exp(1j*k*(X*np.cos(theta) + Y*np.sin(theta))*np.tan(alpha))
    return(uout)

def rect(x):
    y = np.zeros(np.shape(x)).astype(float)
    y[np.abs(x) < 0.5] = 1.0
    y[np.abs(x) == 0.5] = 0.5
    return(y)

def focus(uin,L,lambda_0,zf):
    # converging or diverging phase-front
    # uniform sampling assumed
    # uin - input field
    # L - side length
    # lambda -wavelength
    # zf = focal distance (+ converge, - diverge)
    # uout - output field
    (M,N)=np.shape(uin)
    L = float(L)
    lambda_0 = float(lambda_0)
    zf = float(zf)
    M = float(M)
    N = float(N)
    dx = float(L)/float(M)
    k = 2*np.pi/lambda_0
    x = np.arange(-L/2,L/2,dx)
    (X,Y) = np.meshgrid(x,x)
    return uin*np.exp(-1j*k/(2*zf)*(X**2+Y**2))

L1 = float(0.5)
M=float(250)
dx1=L1/M
x1=np.linspace(-L1/2,L1/2,np.int(L1/dx1))
y1=x1

lambda_0=0.5e-6
k=2*np.pi/lambda_0
w=0.051
z=2000
(X1,Y1)=np.meshgrid(x1,y1)

u1 = rect(X1/(2*w))*rect(Y1/(2*w))

plt.figure(figsize=(4.5,4))
plt.pcolormesh(x1,y1,np.abs(u1**2))

u2 = propFull(tilt(u1,L1,lambda_0,45.0*(np.pi/180.0),0),L1,lambda_0,z)

I2 = np.abs(u2**2)

plt.figure(figsize=(8,7))
plt.pcolormesh(x1,y1,I2)
plt.colorbar()
```
