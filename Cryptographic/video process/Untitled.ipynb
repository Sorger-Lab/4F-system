```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
import imageio
```

```python
for i in range(120):
    screen=np.zeros((500,500))
    x=i%10
    screen[x*50:x*50+50]=255
    imageio.imwrite('images/{0}.png'.format(str(10000+i)),screen.astype(np.uint8))
```

```python
#fourcc = cv2.VideoWriter_fourcc(*'XVID')
fourcc = cv2.VideoWriter_fourcc(*'MJPG')
out = cv2.VideoWriter('output60.avi',fourcc, 60.0, (500,500))
```

```python
for i in range(120):
    screen=np.zeros((500,500))
    x=i%10
    screen[x*50:x*50+50]=255
    out.write(cv2.imread('images/{0}.png'.format(str(10000+i))))
out.release()
```

```python
cap = cv2.VideoCapture('ouput.avi')

while(cap.isOpened()):
    ret, frame = cap.read()

    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    cv2.imshow('frame',gray)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
